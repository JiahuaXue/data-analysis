### Step 4: Machine learning model

#### 4.1 Data processing

```python
# 去掉不需要的列(包括日期型数据）
df_modelling = df.drop(columns=['ID','Dt_Customer', 'Z_CostContact', 'Z_Revenue','Log_Income','Year_Birth'])
df_modelling.head()
```

```python
# one hot encoding, to convert all non-numerical features into numerical features
df_modelling = pd.get_dummies(df_modelling,dtype=int) # 因为不知为何输出了truefalse这种布尔型变量，所以强制指定输出0和1
df_modelling.head().transpose()
```

```python
# Labels and features
x= df_modelling.drop(columns=['Response'])
x.head()
```

```python
y= df_modelling['Response']
y.head()
```

```python
# Test Train Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test = train_test_split(x,y,test_size=0.3,train_size = 0.7)
```

```python
x_train['Income'].count()
```

```python
x_test['Income'].count()
```

#### 4.2 Build machine learning models

```python
# Use the Random Forest to build a machine learning model
from sklearn.ensemble import RandomForestClassifier
```

```python
# define the model to be used for training
clf = RandomForestClassifier(max_depth = 5)
```

```python
# input the training features and label to train the model
clf.fit(x_train, y_train)
```

#### 4.3 Make prediction on test dataset and test the model performance

```python
# prediction
y_predict = clf.predict(x_test)
```

Then we compare the prediction to the y_test to evaluate the accuracy of modelling. We use accuracy, AUC and confusion matrix.

```python
from sklearn.metrics import accuracy_score, precision_score, classification_report, confusion_matrix, roc_curve,auc
```

```python
# Accuracy
accuracy_score(y_test, y_predict)
```

```python
# Precision_score - tp / (tp + fp) ratio 真阳性占预测阳性的比例
precision_score(y_test, y_predict)
```

```python
# AUC, 1 → perfect classification, 0.5 → Random guess
y_prob = clf.predict_proba(x_test)
y_prob
```

```python
pred = y_prob[:,1] # 取第二列的预测概率，即类别1的预测概率

# 计算假阳性率、真阳性率和阈值
fpr, tpr, thresholds = roc_curve(y_test, pred)

#计算AUC
auc(fpr, tpr)
```

```python
# 画ROC curve
import matplotlib.pyplot as plt
plt.plot(fpr,tpr,lw=2)
plt.plot([0,1], [0,1], lw=2, linestyle='--')
```

#### Additional Task

Write an automatic program for parameter tuning Use a for loop to change the random forest max_depth parameter from 3 to 15 and print the AUC for different max_depth. Plot a figure where the x axis is max_depth and y-axis is AUC.

```python
# change the random forest max_depth parameter from 3 to 15 and print the AUC for different max_depth

from sklearn.metrics import roc_auc_score


auc = [] #构建AUC值的list以便画图
min_depth = 3 #避免硬编码让代码更易于修改和维护
max_depth = 16
max_depth_range = range(min_depth, max_depth)  # 不需要 list，range 本身可迭代


for max_depth in max_depth_range:
    
    # 创建模型并训练
    clf = RandomForestClassifier(max_depth = max_depth, random_state=26) #确保模型的随机行为（如样本选择、树构造）一致以便复现结果
    clf.fit(x_train, y_train)

    # 获取预测概率并计算 AUC
    y_prob = clf.predict_proba(x_test)[:,1]
    auc_value = roc_auc_score(y_test, y_prob)
    auc.append(auc_value)

    # 打印结果
    print(f"max_depth:{max_depth}  AUC: {auc_value}")
```

```python
# Plot a figure where the x axis is max_depth and y-axis is AUC
plt.plot(list(max_depth_range),auc)
plt.ylabel("AUC")
plt.xlabel("Max Depth")
plt.show()
```

We can find that the maximum AUC appears when max depth equals to 10.
