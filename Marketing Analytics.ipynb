### Variable transformation

We find the income & Dt_Customer column in object type, so we need to transfer the variable.

```python
print(df.columns.tolist()) # check the name of "Income" column
```

```python
df.columns = df.columns.str.strip() # remove the space of column name
```

```python
df['Income'] = df['Income'].str.replace('[$,]', '', regex=True).astype(float)
```

```python
# Transfer the date variable into datetime format
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True)
df.head()
```

### Deal with null values & outliers

We find null values in income, and we need to transfer the income into numeric varaibles.

```python
df.isnull().sum()
```

A minor portion of the income data contains null values. We try to take a look at the distribution of it first.

```python
sns.catplot(data=df,y="Income", kind="box") # Use boxplot to check the income distribution
```

As the plot show, the income contains outliers at upper side. So we use median value to fill in the null value first, then take log value of income to reduce the ourlier effect.

```python
# fill NA with median income
df['Income'] = df['Income'].fillna(df['Income'].median())
```

```python
# take log value of income
df[df['Income']==0] # Check if we have 0 income to avoid log(0) issue
```

```python
df['Log_Income']=np.log(df['Income'])
df.drop(columns=['Income'], inplace=True)
df.head()
```

```python
# Use IQR(Interquartile Range）to check other outliers
df_filtered = df.loc[:, 'Recency':'NumWebVisitsMonth']
list(df_filtered.columns)
```

```python
for col in list(df_filtered.columns):
    Q1 = df_filtered[col].quantile(0.25)
    Q3 = df_filtered[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df_filtered[col]< lower_bound) | (df_filtered[col]> upper_bound)]
    print(f"Outliers in {col} ：{outliers.shape[0]}")
```

```python
outlier_cols = ["MntWines", "MntFruits", "MntMeatProducts", 
                "MntFishProducts", "MntSweetProducts", "MntGoldProds",
                "NumDealsPurchases", "NumWebPurchases", 
                "NumCatalogPurchases", "NumWebVisitsMonth"]

for col in outlier_cols + ["NumStorePurchases"]:
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.show()
```

The outliers focus on the consumption data, it is usually right skewed, so we don;t delete them. We still do the log transformation on them for the regression.

```python
for col in outlier_cols:
    df[f'log_{col}'] = np.log1p(df[col]) 
    df.drop(columns=[col], inplace=True) #delete the original column
df.head()
```

```python
# PhD v.s. non PhD, t test
from scipy.stats import ttest_ind

PhD = df[df["Education"] == 'PhD']["NumStorePurchases"]
non_PhD = df[df["Education"] != 'PhD']["NumStorePurchases"]

t_stat, p_val = ttest_ind(PhD, non_PhD, equal_var=False)
print(f"T-statistic: {t_stat}, P-value: {p_val}")
```

p<0.05, significant.

```python
# Widow v.s. non Widow, t test
Widow = df[df["Marital_Status"] == 'Widow']["NumStorePurchases"]
non_Widow = df[df["Marital_Status"] != 'Widow']["NumStorePurchases"]

t_stat, p_val = ttest_ind(Widow, non_Widow, equal_var=False)
print(f"T-statistic: {t_stat}, P-value: {p_val}")
```

p>0.05, insignificant.

```python
from statsmodels.stats.proportion import proportions_ztest # Z-test for Proportions

# calculate response number & overall number of people of PhD and non-PhD
phd_responded = df[df['Education'] == 'PhD']['Response'].sum()
phd_total = len(df[df['Education'] == 'PhD'])

non_phd_responded = df[df['Education'] != 'PhD']['Response'].sum()
non_phd_total = len(df[df['Education'] != 'PhD'])

# z-test
count = [phd_responded, non_phd_responded]  
nobs = [phd_total, non_phd_total]  
stat, pval = proportions_ztest(count, nobs)

print(f"Z-score: {stat}, p-value: {pval}")
```

p>0.05, insignificant.

```python
# calculate response number & overall number of people of Widow and non-Widow
Widow_responded = df[df['Marital_Status'] == 'Widow']['Response'].sum()
Widow_total = len(df[df['Marital_Status'] == 'Widow'])

non_Widow_responded = df[df['Marital_Status'] != 'Widow']['Response'].sum()
non_Widow_total = len(df[df['Marital_Status'] != 'Widow'])

# z-test
count = [Widow_responded, non_Widow_responded]  
nobs = [Widow_total, non_Widow_total]  
stat, pval = proportions_ztest(count, nobs)

print(f"Z-score: {stat}, p-value: {pval}")
```

p<0.05, significant.

### Pre-processing of data (One Hot Encoding)

```python
# Pre-processing of data (One Hot Encoding)
df_reg = pd.get_dummies(df,dtype=int,drop_first=True) # delete one category to avoid Dummy Variable Trap in regression
df_reg.head().transpose()
```

### Q1: predict the related factor of store purchase

```python
# train test split
from sklearn.model_selection import train_test_split
```

```python
# OLS model
import statsmodels.api as sm

# Q1: predict the related factor of store purchase
X1 = df_reg.drop(columns=['NumStorePurchases'])
y1 = df_reg["NumStorePurchases"]
X1 = sm.add_constant(X1)

X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.2, random_state=42)

model1 = sm.OLS(y_train, X_train).fit()
print(model1.summary())
```

Significant Variable（p ≤ 0.05）：

Kidhome（-0.4666, p = 0.000）<br>

log_NumWebVisitsMonth（-0.5442, p = 0.001）<br>
Education_Basic（0.6477, p = 0.047）<br>
Age（-0.0101, p = 0.014）<br>

```python
# Use test data to test the model
from sklearn.metrics import mean_squared_error, r2_score

y_pred = model1.predict(X_test)
print(f'R²: {r2_score(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
```

### Q2: US v.s. the Rest of the World in terms of total purchases

```python
from scipy.stats import ttest_ind # Independent two-sample t test

us_purchases = df_reg[df_reg["Country_US"] == 1]["NumStorePurchases"]
row_purchases = df_reg[df_reg["Country_US"] == 0]["NumStorePurchases"]

t_stat, p_val = ttest_ind(us_purchases, row_purchases, equal_var=False)
print(f"T-statistic: {t_stat}, P-value: {p_val}")
```

p >= 0.05 → US fare is not  significantly better than the Rest of the World in terms of total purchases.

### Q3: gold spending v.s. store purchases

```python
# Correlation Analysis
corr = df["log_MntGoldProds"].corr(df["NumStorePurchases"])
print(f"Correlation: {corr}")

# Divide by whether gold spending higher than average amount
gold_high = df[df["log_MntGoldProds"] > df["log_MntGoldProds"].mean()]["NumStorePurchases"]
gold_low = df[df["log_MntGoldProds"] <= df["log_MntGoldProds"].mean()]["NumStorePurchases"]

t_stat, p_val = ttest_ind(gold_high, gold_low, equal_var=False)
print(f"T-statistic: {t_stat}, P-value: {p_val}")
```

From correlation and p<0.05, it supports the statement that people who spent an above average amount on gold in the last years would have more in store purchases.

### Q5: relationship between geographical regional and campaign

```python
# chi squared test
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df["Country"], df["Response"])
chi2, p_val, dof, expected = chi2_contingency(contingency_table)

print(f"Chi2: {chi2}, P-value: {p_val}")
```

p>0.05 → There is not a significant relationship between geographical regional and success of a campaign.
